### Q1 — Exploratory vs. Explanatory
**Prompt:** *"Explain the difference between exploratory and explanatory visualization."*  
**Hints:** Consider audience, purpose, level of polish, and how much iteration you expect.

📝 **Your response:**

- Exploratory visualization: Made for analysts exploring the data to find trends, patterns, or anomalies; quick, iterative, less polished; audience is usually yourself or your team.
- Explanatory visualization: Made for presenting key findings to decision-makers; polished, focused on a clear takeaway or narrative; designed for communication, not discovery.
- Key difference: Exploratory asks “What’s in the data?” while explanatory tells “Here’s what you need to know from the data.”

### Q2 — Chart Types & When to Use Them
**Prompt:** *"Describe 5 common chart types (e.g., bar, line, scatter, histogram, map) and the type of data analysis question each is best suited to answer."*  

For each chart, list: **Question archetype**, **Data shape**, **Pitfalls**, **Alternatives**.

📝 **Your response (template):**
1. **Bar:**  
   - Question archetype: Compare categories (Which product sells the most?).

   - Data shape: Categorical × numerical.

   - Pitfalls: Too many bars → clutter; inconsistent scales can mislead.

   - Alternatives: Column chart, stacked bar for composition.
2. **Line:**  
   - Question archetype: Show trends over time (How do sales change month to month?).

   - Data shape: Time series, continuous × numerical.

   - Pitfalls: Uneven time intervals distort trends; too many lines → hard to read.

   - Alternatives: Area chart for emphasis, sparklines for small multiples.
3. **Scatter:**  
   - Question archetype: Relationship/correlation (Is price related to demand?).

   - Data shape: Two numerical variables (optionally add color/size for more dimensions).

   - Pitfalls: Overplotting with large datasets; correlation ≠ causation.

   - Alternatives: Bubble chart (adds size), hexbin/density plot for large data.
4. **Histogram:**  
   - Question archetype: Distribution (How are order sizes distributed?).

   - Data shape: Single numerical variable.

   - Pitfalls: Bin width choice can hide or exaggerate patterns; misleading if unequal bins.

   - Alternatives: Density plot, box plot for summary.
5. **Map:**  
   - Question archetype: Geospatial pattern (Where are most sales located?).

   - Data shape: Geographic (lat/long, state, region) × numerical/categorical.

   - Pitfalls: Distorted if area ≠ value; choropleths can mislead when regions differ in size/population.

   - Alternatives: Proportional symbol maps, heatmaps, small multiples.

### Create a Synthetic Airline Dataset (if you don't have one)
Use this dataset to prototype visualizations and dashboards. It contains realistic columns for the KPIs you defined. You can always swap in your BigQuery data later.


````python
# Generate a synthetic airline dataset
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

np.random.seed(467)

dates = pd.date_range(start='2024-01-01', end='2024-06-30', freq='D')
carriers = ['AA', 'DL', 'UA', 'WN', 'AS']
origins = ['ATL','DFW','DEN','ORD','LAX','SEA','SFO','PHX','CLT','MCO']
dests =   ['JFK','LGA','BOS','SFO','LAX','SEA','MIA','IAH','IAD','PHL']

rows = []
for d in dates:
    for carrier in carriers:
        # number of flights per day per carrier
        n = np.random.randint(80, 140)
        for _ in range(n):
            o = np.random.choice(origins)
            de = np.random.choice(dests)
            if o == de:
                continue
            distance = np.random.randint(300, 2800)
            seats = np.random.choice([90,120,150,180,210])
            load_factor = np.clip(np.random.normal(0.82, 0.07), 0.55, 0.98)
            pax = int(seats * load_factor)
            dep_delay = int(np.random.normal(8, 15))
            arr_delay = dep_delay + int(np.random.normal(0, 12))
            dep_delay = max(dep_delay, -5)
            arr_delay = max(arr_delay, -10)
            cancelled = np.random.rand() < 0.025
            diverted = np.random.rand() < 0.005
            revenue = pax * np.random.uniform(120, 420)
            casm = np.random.uniform(0.08, 0.16)  # cost per available seat mile
            cost = casm * seats * distance
            fuel_burn = distance * np.random.uniform(3.0, 4.5)  # gallons per seat-mile proxy
            nps = np.random.normal(35, 15)  # if you want a customer metric proxy
            rows.append({
                'date': d,
                'carrier': carrier,
                'origin': o,
                'dest': de,
                'distance_miles': distance,
                'seats': seats,
                'passengers': pax,
                'dep_delay_min': dep_delay,
                'arr_delay_min': arr_delay,
                'cancelled': int(cancelled),
                'diverted': int(diverted),
                'revenue_usd': round(revenue, 2),
                'cost_usd': round(cost, 2),
                'fuel_burn_gal': round(fuel_burn, 1),
                'nps': round(nps, 1)
            })

df_flights = pd.DataFrame(rows)

# Derived KPIs
df_flights['on_time'] = (df_flights['arr_delay_min'] <= 15).astype(int)
df_flights['load_factor'] = (df_flights['passengers'] / df_flights['seats']).round(3)
df_flights['yield_usd_per_pax_mile'] = (df_flights['revenue_usd'] / (df_flights['passengers'] * df_flights['distance_miles']).clip(lower=1)).round(4)
df_flights['asm'] = df_flights['seats'] * df_flights['distance_miles']  # available seat miles
df_flights['casm'] = (df_flights['cost_usd'] / df_flights['asm']).round(4)
df_flights['month'] = df_flights['date'].dt.to_period('M').astype(str)

df_flights.head()
````

Given `df_flights` with `cost_usd`, `revenue_usd`, and `casm`, produce a **scatter plot** of cost vs. revenue with points colored by `casm` (continuous colormap) and add a diagonal y=x reference line. Include labels and legend.

````python
# Prompt A — On-Time Rate by Carrier (Bar Chart)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Compute on-time rate, convert to percentage, and sort
on_time_rate = (
    df_flights.groupby("carrier")["on_time"]
    .mean()
    .mul(100)
    .sort_values(ascending=False)
    .reset_index(name="on_time_rate")
)

# Plot
plt.figure(figsize=(8, 5))
sns.set_style("whitegrid")
# Assign 'carrier' to hue to avoid the FutureWarning
barplot = sns.barplot(
    data=on_time_rate,
    x="carrier",
    y="on_time_rate",
    hue="carrier",
    palette="Blues_d",
    legend=False
)

# Add percentage labels on bars
for p in barplot.patches:
    barplot.annotate(
        f"{p.get_height():.1f}%",
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha="center", va="bottom",
        fontsize=10, color="black", xytext=(0, 3),
        textcoords="offset points"
    )

# Titles and labels
plt.title("On-Time Rate by Carrier", fontsize=14, weight="bold")
plt.ylabel("On-Time Rate (%)")
plt.xlabel("Carrier")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

````python
# Prompt B — Average Arrival Delay over Time (Line Chart)
import pandas as pd
import matplotlib.pyplot as plt

# Ensure date is datetime
df_flights["date"] = pd.to_datetime(df_flights["date"])

# Aggregate: daily average arrival delay
daily_delay = (
    df_flights.groupby("date")["arr_delay_min"]
    .mean()
    .reset_index()
    .rename(columns={"arr_delay_min": "avg_delay"})
)

# Compute 7-day rolling average
daily_delay["rolling_7d"] = daily_delay["avg_delay"].rolling(7, center=True).mean()

# Plot
plt.figure(figsize=(10, 6))
plt.plot(daily_delay["date"], daily_delay["avg_delay"], label="Daily Avg Delay", alpha=0.6)
plt.plot(daily_delay["date"], daily_delay["rolling_7d"], label="7-Day Rolling Avg", color="red", linewidth=2)

# Labels and styling
plt.title("Daily Average Arrival Delay with 7-Day Rolling Average", fontsize=14, weight="bold")
plt.xlabel("Date", fontsize=12)
plt.ylabel("Average Arrival Delay (minutes)", fontsize=12)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
````

````python
#Prompt C — Delay Distribution (Histogram)
import matplotlib.pyplot as plt
import numpy as np

# Compute percent of flights within [-5, 15] minutes
within_window = ((df_flights["arr_delay_min"] >= -5) & (df_flights["arr_delay_min"] <= 15)).mean() * 100

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(df_flights["arr_delay_min"], bins=50, color="skyblue", edgecolor="black", alpha=0.7)

# Add vertical reference line at 0
plt.axvline(0, color="red", linestyle="--", linewidth=2, label="On-time (0 min)")

# Annotate percent within [-5, 15]
plt.text(
    0.95, 0.95,
    f"{within_window:.1f}% within [-5, 15] min",
    transform=plt.gca().transAxes,
    ha="right", va="top",
    fontsize=12, bbox=dict(facecolor="white", edgecolor="black", boxstyle="round,pad=0.3")
)

# Labels and styling
plt.title("Distribution of Arrival Delays", fontsize=14, weight="bold")
plt.xlabel("Arrival Delay (minutes)", fontsize=12)
plt.ylabel("Number of Flights", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
````

````python
# Prompt D — Load Factor by Route (Top 15) (Horizontal Bar)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a route column
df_flights["route"] = df_flights["origin"] + " → " + df_flights["dest"]

# Compute mean load factor by route
route_load = (
    df_flights.groupby("route")["load_factor"]
    .mean()
    .reset_index()
    .sort_values("load_factor", ascending=False)
    .head(15)
)

# Plot horizontal bar chart
plt.figure(figsize=(10, 8))
sns.set_style("whitegrid")
barplot = sns.barplot(
    data=route_load,
    x="load_factor",
    y="route",
    palette="Blues_r",
    hue="route",
    legend=False
)

# Labels and title
plt.title("Top 15 Routes by Average Load Factor", fontsize=14, weight="bold")
plt.xlabel("Average Load Factor")
plt.ylabel("Route")

# Add value labels
for p in barplot.patches:
    plt.text(
        p.get_width() + 0.005,                # x position
        p.get_y() + p.get_height() / 2,     # y position
        f"{p.get_width():.3f}",             # formatted value
        va="center"
    )

plt.xlim(right=route_load['load_factor'].max() * 1.05) # Adjust x-axis limit for labels
plt.tight_layout()
plt.show()
````

````python
# Prompt E — Cost vs. Revenue with CASM Coloring (Scatter)
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.set_style("whitegrid")

# Scatter plot: revenue vs cost, colored by CASM
scatter = plt.scatter(
    df_flights["cost_usd"],
    df_flights["revenue_usd"],
    c=df_flights["casm"],
    cmap="viridis",
    alpha=0.7,
    edgecolor="k"
)

# Diagonal reference line y = x
lims = [
    min(df_flights["cost_usd"].min(), df_flights["revenue_usd"].min()),
    max(df_flights["cost_usd"].max(), df_flights["revenue_usd"].max())
]
plt.plot(lims, lims, 'r--', linewidth=2, label="Break-even (Cost = Revenue)")

# Labels, title, legend
plt.xlabel("Cost (USD)", fontsize=12)
plt.ylabel("Revenue (USD)", fontsize=12)
plt.title("Cost vs. Revenue Colored by CASM", fontsize=14, weight="bold")
plt.colorbar(scatter, label="CASM (Cost per Available Seat Mile)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
````
